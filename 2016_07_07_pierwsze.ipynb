### Jedna cząstka 2D

````python
liczba_iteracji = int(1e5)
polozenia = np.zeros((liczba_iteracji, 2))

# przesuniecia na kazdej osi od -0.5 do +0.5
for i in range(1, liczba_iteracji):
    przesuniecie = np.random.random(2)-0.5
    polozenia[i] = polozenia[i-1] + przesuniecie
x = polozenia[:, 0]
y = polozenia[:, 1]
plt.plot(*polozenia.T)
````

# Kilka cząstek, 2D, najprościej

````python
liczba_iteracji = int(1e3)
liczba_czastek = 2
for czastka in range(liczba_czastek):
    polozenia = np.zeros((liczba_iteracji, 2))
    # przesuniecia na kazdej osi od -0.5 do +0.5
    for i in range(1, liczba_iteracji):
        przesuniecie = np.random.random(2)-0.5
        polozenia[i] = polozenia[i-1] + przesuniecie
    x = polozenia[:, 0]
    y = polozenia[:, 1]
    plt.plot(*polozenia.T, alpha=0.7)
````

### Kilka cząstek, wektorowo (szybko)

````python
liczba_iteracji = int(1e2)
liczba_czastek = 200
liczba_kolumn = 2*liczba_czastek

polozenia = np.zeros((liczba_iteracji, liczba_kolumn))
# przesuniecia na kazdej osi od -0.5 do +0.5
for i in range(1, liczba_iteracji):
    przesuniecie = np.random.random(liczba_kolumn)-0.5
    polozenia[i] = polozenia[i-1] + przesuniecie
x = polozenia[:, 0::2]
y = polozenia[:, 1::2]
plt.plot(x, y, ".", alpha=0.3);
````

### Random walk 3D

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
liczba_iteracji = int(1e4)                      #tutaj tez mniej                            
liczba_czastek = 100                              #tymczasowo mniej
liczba_kolumn = 3*liczba_czastek                # 3 zamiast 2

polozenia = np.zeros((liczba_iteracji, liczba_kolumn))
# przesuniecia na kazdej osi od -0.5 do +0.5
for i in range(1, liczba_iteracji):
    przesuniecie = np.random.random(liczba_kolumn)-0.5
    polozenia[i] = polozenia[i-1] + przesuniecie

#dodajemy plot 3d
fig = plt.figure(figsize=(6, 8))
ax = fig.add_subplot(111, projection='3d')
for i in range(0, liczba_czastek):
    x = polozenia[:, 0+3*i]       #3 zamiast 2
    y = polozenia[:, 1+3*i]       # takooż
    z = polozenia[:, 2+3*i]
    ax.plot(x, y, z);
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
### magiczna linijka o której zawsze zapominam, a która pozwala obracać wewnątrz notebooka (ale już nie animować)
### przetestujcie na włączonym notebooku (zapisywać można z tego co wiem tylko jako statyczne obrazki)
liczba_iteracji = int(1e3)                                #tutaj tez mniej                            
liczba_czastek = 3                              #tymczasowo mniej
liczba_kolumn = 3*liczba_czastek                # 3 zamiast 2

polozenia = np.zeros((liczba_iteracji, liczba_kolumn))
# przesuniecia na kazdej osi od -0.5 do +0.5
for i in range(1, liczba_iteracji):
    przesuniecie = np.random.random(liczba_kolumn)-0.5
    polozenia[i] = polozenia[i-1] + przesuniecie

#dodajemy plot 3d
fig = plt.figure(figsize=(6, 8))
ax = fig.add_subplot(111, projection='3d')
for i in range(0, liczba_czastek):
    x = polozenia[:, 0+3*i]       #3 zamiast 2
    y = polozenia[:, 1+3*i]       # takooż
    z = polozenia[:, 2+3*i]
    ax.plot(x, y, z);
plt.show()
````

### Fitowanie krzywej
[Średnia odległość cząstki od punktu początkowego powinna lecieć jak](https://en.wikipedia.org/wiki/Brownian_motion#Einstein.27s_Theory) $r \sim \sqrt{t}$

````python
polozenia.shape
````

````python
srednie_odleglosci = (polozenia**2).sum(axis=1)**0.5
%matplotlib inline
plt.plot(srednie_odleglosci)
````

````python
t = np.arange(liczba_iteracji)

def pierwiastek(x, A, b):
    return A*np.sqrt(b*x)

from scipy.optimize import curve_fit
# curve_fit?
parametry, kowariancje = curve_fit(pierwiastek, t, srednie_odleglosci)

plt.plot(t, srednie_odleglosci)
plt.plot(t, pierwiastek(t, *parametry), "--")
````
